```python
import numpy as np 
import matplotlib.pyplot as plt

n = 1000 
df = 10
x = np.zeros(n)

for i in range(n):
    exp_dists = np.random.exponential(2, size=int(df/2))
    chi_sq = np.sum(exp_dists)
    x[i] = chi_sq
```

```python
mean = np.mean(x)
std = np.std(x)

plt.hist(x, bins=20)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

a = 6
b = 4 
result = minimize_scalar(lambda x : -stats.beta.pdf(x, a, b),
                         bounds=(0,1), method='bounded')
M = -result.fun

n = 3000
y = np.random.uniform(size=n)
u = np.random.uniform(size=n)
fy = stats.beta.pdf(y,a, b)
gy = 1

x = y[u < (fy/(M + gy))]
x = x[:1000]
plt.hist(x, bins=20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import  poisson

lambda_ = 0.3
n = 10

# Calc the probabilities of y
probs = [poisson.cdf(k, lambda_)**n - poisson.cdf(k-1, lambda_)**n for k in range(11)]

plt.bar(range(11), probs)
plt.xlabel('Maximum number of earthquakes per year for the period of 10 years')
plt.ylabel('Probability')
plt.title('Distribution of the maximum number of earthquakes per year for the period of 10 years')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
N = 1000 
n = 10
samples = np.random.uniform(0, 1, size=(N, n)) 
sample_means = samples.mean(axis=1)
# Parameters for the normal distribution according to the CLT 
mean = 0.5
var = 1/12 / n
std = np.sqrt(var)
plt.hist(sample_means, bins=20, density=True, alpha=0.6, color='b', label='Sample means')
x = np.linspace(mean - 4 * std, mean + 4 * std, 100)
y = norm.pdf(x, mean, std)
plt.plot(x, y, 'r', label='Expected normal distribution')
plt.xlabel('Sample mean')
plt.ylabel('Density')
plt.title('Histogram of 1000 sample means')
plt.legend(loc = 'upper right', fontsize = 8)
plt.show()
```

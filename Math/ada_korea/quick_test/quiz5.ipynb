```python
import numpy as np
T = np.array([[0.6, 0.7], [0.4, 0.3]]) 
threshold = 0.00005
n = 10
p_0 = np.array([1, 0])
p_n = np.matmul(T, p_0)
step_counter = 1

for i in range(2, n+1):
    print(p_n)
    prev = p_n
    p_n = np.matmul(T, prev)
    if abs(p_n[0]-prev[0]) < threshold:
        break 
    step_counter += 1

print(step_counter)
print(p_n)
```

```python
import numpy as np
import matplotlib.pyplot as plt
N = 1000
k=5
n = 10
p = 18 / 38
durations = np.zeros(N)

for i in range(N): 
    capital = k
    duration = 0
    while (capital > 0 and capital < n):
        bet_result = np.random.uniform() < p
        capital += 1 if bet_result else -1
        duration += 1
    durations[i] = duration

plt.hist(durations, color="lightblue", bins=20) 
plt.xlabel("Duration (T)")
plt.ylabel("Frequency")
plt.title("Duration of the Game")
plt.show()
mean_duration = np.mean(durations)
std_duration = np.std(durations)
print("Mean duration:", mean_duration)
print("Standard deviation of duration:", std_duration)
```
